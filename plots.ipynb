```python
validation_accuracy = []
train_accuracy = []

with open('losscopy/loss_gelu_quora.txt', 'r') as file:
    for line in file:
        columns = line.strip().split()
        if len(columns) >= 4:
            try:
                c_value = float(columns[2])
                d_value = float(columns[3])
                validation_accuracy.append(d_value)
                train_accuracy.append(c_value)
            except ValueError:
                continue

print("Validation Accuracy:", validation_accuracy)
print("Train Accuracy:", train_accuracy)
```
